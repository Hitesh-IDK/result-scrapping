```python
import requests
import time

from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService 
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait

import subprocess
import webbrowser
```

```python
#Results URL
url = 'https://results.vtu.ac.in/JJEcbcs23/index.php'

#Driver for chrome
driver = webdriver.Chrome(service=ChromeService(
    ChromeDriverManager().install()))

driver.get(url)
```

```python
#Other wbsite driver
myUrl = 'http://localhost:8000/Captcha%20UI/'
solver = webdriver.Chrome(service=ChromeService(
    ChromeDriverManager().install()))

def initialize_solver():
    solver.get(myUrl)
```

```python
def usn_input(usn):
    #Get USN input element selected
    usn_element = driver.find_element(By.NAME, 'lns')
    
    #Manipulate USN Input
    usn_element.clear()
    usn_element.send_keys(usn)
```

```python
def save_captcha():
    #Get the captcha element ready
    captcha_box = driver.find_element(By.NAME, 'captchacode')

    #Get captcha image and save
    image = driver.find_element(By.XPATH, '//img[@alt="CAPTCHA code"]')
    with open ('C://Users//10hit//Projects//Untitled Folder//Captcha UI//public//captcha.png', 'wb') as captcha:
        captcha.write(image.screenshot_as_png)
```

```python
def get_captcha_answer():
    #Get captcha answer from user
    while 1:
        captcha_input_element = solver.find_element(By.CLASS_NAME, 'captcha__input')
        captcha_input_ready = captcha_input_element.get_attribute('data-isready')

        #If captcha read by user
        if captcha_input_ready == 'True':
            break
            
    captcha_answer = captcha_input_element.get_attribute('value')
    return captcha_answer
```

```python
def send_captcha(answer):
    #Get the captcha element ready
    captcha_box = driver.find_element(By.NAME, 'captchacode')
    
    #Send input to captcha box
    captcha_box.clear()
    captcha_box.send_keys(answer)
```

```python
def validate_captcha():
    submit = driver.find_element(By.ID, 'submit')
    submit.click()
    
    try:
        alert = driver.switch_to.alert
    except:
        #No error
        return True
    else:
        #Error
        alert.accept()
        return False
    
```

```python
def sleepTillReady():
    time.sleep(3)
```

```python
#limits to regular and lateral students
usn_limit_reg = 47
usn_limit_lat = 16

usns_reg = []
usns_lat = []

for i in range(usn_limit_reg):
    if i + 1 == 25:
        continue
    usns_reg.append(f"4VM21IS0{str(i + 1).rjust(2, '0')}")
    
for i in range(usn_limit_lat):
    usns_lat.append(f"4VM22IS4{str(i).rjust(2, '0')}")
```

```python
# with open('data.csv', 'w') as file:
#     file.write("USN,Student Name,Subject Code,Subject Name,Internal Marks,External Marks,Total,Result\n")

def extractData():
    breakTags = driver.find_elements(By.TAG_NAME, "b")
    tdTags = driver.find_elements(By.TAG_NAME, "td")
    subjectData = driver.find_element(By.CLASS_NAME, 'divTableBody')
        
    student = {
        "name": '',
        "usn": ''
    }
        
    printNext = False

    for element in tdTags:
        if printNext:
            if element.text.strip() != ':':
                textData = element.text[2:]
                if textData[:3] == '4VM':
                    student["usn"] = textData
                else:
                    student["name"] = textData
            else:
                print('Wrong: ', element.text)
            printNext = False

        if element.text.strip() == 'Student Name':
            printNext = True

        if element.text.strip() == 'University Seat Number':
            printNext = True

    print('SUBJECT DATATATATATATT')
    for i, item in enumerate(subjectData.text.split('\n')):
        splitItem = item.split(' ')
        if i == 0:
            continue
#             print(splitItem[:2])
#             print(splitItem[2:4])
#             print(splitItem[4:6])
#             print(splitItem[6:8])
#             print(splitItem[8])
#             print(splitItem[9:])
        else:
            reverseSplitItem = splitItem[::-1]
            with open('data.csv', 'a') as data:
                data.write(f'{student["usn"]},{student["name"]},{reverseSplitItem[-1]},{" ".join(reverseSplitItem[5:-1]).replace(",", " -")},{reverseSplitItem[4]},{reverseSplitItem[3]},{reverseSplitItem[2]},{reverseSplitItem[1]}\n')
    
    with open('data.csv', 'a') as data:
        data.write(',,,,,,,\n')
```

```python


#First usn is hard coded, to ensure captcha is up to date
#Get Captcha ready
save_captcha()

#Send input to usn box from here
usn_input('4VM21IS001')

# #Initialize solver (captcha solving server)
initialize_solver()
time.sleep(3)

solver.refresh()
solver.execute_script(f'document.querySelector(".usn").textContent = "USN - 4VM21IS001";')
solver.maximize_window()

#Get captcha answer from user
captcha_answer = get_captcha_answer()

#Send input to captcha box
send_captcha(captcha_answer)

#check if correct captcha
while not validate_captcha():
    #validation failed, restart
    #Get Captcha ready
    save_captcha()
    
    #Send input to usn box from here
    usn_input('4VM21IS001')
    
    sleepTillReady()
    solver.refresh()
    
    #Get captcha answer from user
    try:
        captcha_answer = get_captcha_answer()
    except:
        captcha_answer = 'qqqqqq'
    
    #Send input to captcha box
    send_captcha(captcha_answer)
    
#Extract data from the page
extractData()

#Go back to result page
driver.back()

#sleep for a while
time.sleep(3)

#repeat for the rest - regular students
for usn in usns_reg:
    #skip first usn
    if usn == usns_reg[0]:
        continue
    
    #Get Captcha ready
    save_captcha()

    #Send input to usn box from here
    usn_input(usn)
    
    solver.refresh()
    solver.execute_script(f'document.querySelector(".usn").textContent = "USN - {usn}";')
    #Get captcha answer from user
    captcha_answer = get_captcha_answer()

    #Send input to captcha box
    send_captcha(captcha_answer)

    #check if correct captcha
    while not validate_captcha():
        #validation failed, restart
        #Get Captcha ready
        save_captcha()

        #Send input to usn box from here
        usn_input(usn)

        sleepTillReady()
        solver.refresh()

        #Get captcha answer from user
        try:
            captcha_answer = get_captcha_answer()
        except:
            captcha_answer = 'qqqqqq'

        #Send input to captcha box
        send_captcha(captcha_answer)

    #Extract data from the page
    extractData()

    #Go back to result page
    driver.back()

    #sleep for a while
    time.sleep(3)

start = False   
#repeat for the rest - lateral students
for usn in usns_lat:
    #skip first usn
    if usn == usns_lat[0]:
        continue
        
    if usn == '4VM22IS411':
        start = True
    
    if not start:
        continue
    
    #Get Captcha ready
    save_captcha()

    #Send input to usn box from here
    usn_input(usn)

    solver.refresh()
    solver.execute_script(f'document.querySelector(".usn").textContent = "USN - {usn}";')
    #Get captcha answer from user
    captcha_answer = get_captcha_answer()

    #Send input to captcha box
    send_captcha(captcha_answer)

    #check if correct captcha
    while not validate_captcha():
        #validation failed, restart
        #Get Captcha ready
        save_captcha()

        #Send input to usn box from here
        usn_input(usn)

        sleepTillReady()
        solver.refresh()

        #Get captcha answer from user
        try:
            captcha_answer = get_captcha_answer()
        except:
            captcha_answer = 'qqqqqq'

        #Send input to captcha box
        send_captcha(captcha_answer)

    #Extract data from the page
    extractData()

    #Go back to result page
    driver.back()

    #sleep for a while
    time.sleep(3)
    
driver.close()
solver.close()
```
